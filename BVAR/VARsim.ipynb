# Example Script for the Verctor Autoregression Simulation using VARsim


```python
from VARsim import *
import numpy as np
```

```python
n_obs = 100
simulator = VARSimulator(n_obs=n_obs, coefficients=[
                         [0.8, -0.2], [0.3, 0.5]], initial_values=[[1], [2]], noise_stddev=1.0, output='show')
simulator.run("VARsim")
print(simulator.data)
```

```python
# If no coefficients are specified, the default values are used
simulator = VARSimulator(n_obs=n_obs, initial_values=[
                         [1], [2]], noise_stddev=4.0, output='show')
simulator.run("VARsim")
simulator.print_parameters()
```

```python
# Simulating a multivariate VAR(1) process

n_obs = 100
nX = 2
A = np.array([[0.8, -0.2], [0.3, 0.5]])
X0 = np.array([1, 2]).reshape((2, 1))
nS = 3
# B = np.array([[0, 0], [0.3, 0.5]])
B = np.zeros((nS, nX))
B[0, 1] = 0.8
B[2, 1] = -0.5
# print(B)
S0 = np.zeros(nS).reshape((nS, 1))
simulator = VARSimulator(n_obs=n_obs, coefficients=A,
                         initial_values=X0, noise_stddev=4.0, output='show')
simulator.run("MVARsim", coefficientsM=B, initial_valuesM=S0)

simulator.print_parameters()
```

```python
n_obs = 100
nX = 2
A = np.array([[0.8, -0.2], [0.3, 0.5]])
X0 = np.array([1, 2]).reshape((2, 1))
nS = 3
# B = np.array([[0, 0], [0.3, 0.5]])
B = np.zeros((nS, nX))
B[0, 1] = 0.8
B[2, 1] = -0.5
# print(B)
S0 = np.zeros(nS).reshape((nS, 1))
simulator = VARSimulator(n_obs=n_obs, coefficients=A,
                         initial_values=X0, noise_stddev=4.0, output='none')
simulator.run("MVARsim", coefficientsM=B, initial_valuesM=S0)

# Call the make_plot_stacked() method
make_plot_stacked(simulator.data, simulator.dataM)
```
